```python
#Januar  Komerc. oznaka  do prvega /
#Potreben encoding utf-16
znamke2014 = {}
znamkemodel2014 = {}
prodaja2014 = 0
pravne2014 = 0
fizicnem2014 = 0
fizicnez2014 = 0
reader = DictReader(open('podatki/2014/NIO_podatki_vozila_prve_reg_januar_2014.csv','rt',encoding='utf-16'), delimiter=';')
for row in reader:
    #print(row.keys())
    if row['J-Kategorija in vrsta vozila (opis)'].lower() == 'osebni avtomobil' and row['B-Datum prve registracije vozila'] == row['2A-Datum prve registracije vozila v SLO'] and row['X-Oblika nadgradnje (opis)'].lower() == 'vozilo z dvižnimi vrati zadaj' and float(row['P.1.2-Nazivna moc'].replace(",", ".")) >= 180:
        if row['Komerc. oznaka  do prvega /'].upper() in modelip:
            continue
        prodaja2014 += 1
        if row['D.1-Znamka'].upper() in znamke2014.keys():
            znamke2014[row['D.1-Znamka'].upper()] += 1
        else:
            znamke2014[row['D.1-Znamka'].upper()] = 1
        if row['D.1-Znamka'].upper() + ' ' + row['Komerc. oznaka  do prvega /'].upper() in znamkemodel2014.keys():
            znamkemodel2014[row['D.1-Znamka'].upper() + ' ' + row['Komerc. oznaka  do prvega /'].upper()] += 1
        else:
            znamkemodel2014[row['D.1-Znamka'].upper() + ' ' + row['Komerc. oznaka  do prvega /'].upper()] = 1
        if row['C-Ali je uporabnik pravna ali fizicna oseba'].upper() == 'P':
            pravne2014 += 1
        elif row['C-Spol uporabnika (ce gre za fizicno osebo)'].upper() == 'Z':
            fizicnez2014 += 1
        else:
            fizicnem2014 += 1
            
```
