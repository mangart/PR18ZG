**<h1>Leto 2012</h1>**

```python
vozila = ['SUZUKI','PEUGEOT','CITROEN','RENAULT','VOLKSWAGEN','OPEL','HONDA','SEAT','BMW','FORD','MERCEDES BENZ','ŠKODA','MITSUBISHI','AUDI','HYUNDAI','KIA','FIAT','TOYOTA','MAZDA','NISSAN','DACIA']
from csv import DictReader
znamke = {}
reader = DictReader(open('podatki/2012/Vozila2012-1stRegs.csv','rt'), delimiter=';')
for row in reader:
    #print(row.keys())
    if row['Kategorija vozila - opis'].lower() == 'osebni avtomobil' and row['Datum prve registracije'] == row['Datum prve registracije Slo']:
        if row['Znamka'].upper() in znamke.keys():
            znamke[row['Znamka'].upper()] += 1
        else:
            znamke[row['Znamka'].upper()] = 1
```

```python
znamke
```

```python
% matplotlib inline
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
plt.style.use('PR.mplstyle')
```

```python
fig = plt.figure(figsize=(15,13))
plt.bar(range(len(znamke)), znamke.values())
plt.xticks(range(len(znamke)), list(znamke.keys()),rotation='vertical')
plt.title("Število registracij vozil po znamkah za leto 2012")
plt.show()
```

```python
znamke1 = {}
for i in vozila:
    if znamke[i] >= 100:
        znamke1[i] = znamke[i]
```

```python
znamke1
```

```python
fig = plt.figure(figsize=(15,13))
plt.bar(range(len(znamke1)), znamke1.values())
plt.xticks(range(len(znamke1)), list(znamke1.keys()),rotation='vertical')
plt.title("Število registracij vozil po znamkah za leto 2012")
plt.show()
```
