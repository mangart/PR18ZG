**<h1>Registracije vozil po Krajevnih enotah(oznake na tablici)</h1>**

**<h1>Leto 2012</h1>**

```python

from csv import DictReader
zacasniKraji = {}
kraji = {}
noviKraji = {}
casNoviKraji = {}
stevec = 0
tevec = 0
reader = DictReader(open('podatki/2012/Vozila2012-1stRegs.csv','rt'), delimiter=';')
for row in reader:
    #print(row.keys())
    if row['Kategorija vozila - opis'].lower() == 'osebni avtomobil':
        stevec += 1
        if row['Reg območje oznaka na tablici'].upper() in zacasniKraji.keys():
            zacasniKraji[row['Reg območje oznaka na tablici'].upper()] += 1
        else: 
            zacasniKraji[row['Reg območje oznaka na tablici'].upper()] = 1
        if row['Datum prve registracije'] == row['Datum prve registracije Slo']:
            tevec += 1
            if row['Reg območje oznaka na tablici'].upper() in casNoviKraji.keys():
                casNoviKraji[row['Reg območje oznaka na tablici'].upper()] += 1
            else:
                casNoviKraji[row['Reg območje oznaka na tablici'].upper()] = 1
```

```python
for i in zacasniKraji.keys():
    zacasniKraji[i] = zacasniKraji[i] / stevec

for i in casNoviKraji.keys():
    casNoviKraji[i] = casNoviKraji[i] / tevec
```

```python
import copy
```

```python
kraji['2012'] = copy.deepcopy(zacasniKraji)
noviKraji['2012'] = copy.deepcopy(casNoviKraji)
```
