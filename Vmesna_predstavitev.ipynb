### 3. Malenkotni rezultati do sedaj

```python
vozila = {}
for i in range(len(data)):
    if data[i][8] in vozila.keys():
        vozila[data[i][8]] += 1
    else:
        vozila[data[i][8]] = 1
```

```python
vozila
```

```python
fig = plt.figure(figsize=(15,13))
plt.bar(range(len(vozila)), vozila.values())
plt.xticks(range(len(vozila)), list(vozila.keys()),rotation='vertical')
plt.title("Število registracij vozil po znamkah za leto 2012")
plt.show()
```

```python
vozila1 = {}
for i in vozila.keys():
    if vozila[i] >= 1000:
        vozila1[i] = vozila[i]
```
